```python
#announce color and list of iris types for scatter plot
color=["red","green","blue"]
categories=["Iris-setosa","Iris-versicolor","Iris-virginica"]
```

```python
#Use while loop to iterate through the dataset
for i in range(3):
    x=df[df['Category']==categories[i]]
    plt.scatter(x["Sepal Length (cm)"],x["Sepal Width (cm)"], c=color[i], label=categories[i])
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
```

```python
#Use while loop to iterate through the dataset 
for j in range(3):
    y=df[df["Category"]==categories[j]]
    plt.scatter(y["Petal Length (cm)"],y["Petal Width (cm)"], c=color[j], label=categories[j])
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend()
```

```python
sns.set()
sns.pairplot(df[["Sepal Length (cm)","Sepal Width (cm)", "Petal Length (cm)", "Petal Width (cm)", "Category"]], hue="Category", diag_kind="kde" )
```

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
# Separate features and target  
data = df.values
X = data[:,0:4]
Y = data[:,4]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
# Support vector machine algorithm
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, y_train)
```

```python
predictions = svn.predict(X_test)

# Showing the accuracy score of the projected classifications, which is determined by the accuracy_score() function
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
X_new = np.array([[5.4,3.7,1.5,0.2], [  6.3,3.4,5.6,2.4 ], [  5.6,2.6,3.5,1.0]])
#Prediction of the species from the input vector
prediction = svn.predict(X_new)
print("Prediction of Species: {}".format(prediction))
```

```python
# Save the model
import pickle
with open('SVM.pickle', 'wb') as f:
    pickle.dump(svn, f)

# Load the model
with open('SVM.pickle', 'rb') as f:
    model = pickle.load(f)
model.predict(X_new)
```
